```python
import pandas as pd
import numpy as np
from pathlib import Path

path = Path("../data/raw/tmdb_6000")

files = [
    "tmdb_6000_movie_dataset.csv",
    "tmdb_6000_movie_credits.csv",
    "tmdb_6000_movie_ratings.csv"
]

for f in files:
    print("\n==============================")
    print("FILE:", f)
    print("==============================")
    df = pd.read_csv(path / f, nrows=200)   
    
    print("Shape (first 200 rows):", df.shape)
    print("\nColumns:")
    print(df.columns.tolist())
    
    print("\nHead (2 rows):")
    display(df.head(2))
    
    print("\nMissing values (first 200 rows):")
    print(df.isna().sum().sort_values(ascending=False).head(10))
    
```
